```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = np.arange(10,30,0.13)
y = 2.222*X + 3.333
```

```python
split = int(len(X)*0.8)
X_train,y_train = X[:split],y[:split]
X_test,y_test = X[split:],y[split:]
```

```python
X_train
```

```python
len(X_train)
```

```python
def plot_data(X_train,y_train,X_test,y_test,predictions=None):
    plt.scatter(X_train,y_train,c='b',label='Training data')
    plt.scatter(X_test,y_test,c='r',label='Testing')
    if predictions is not None:
        plt.plot(X_test,predictions,c='g',label='Predictions')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('X vs y')
    plt.legend()
    plt.show()
```

```python
plot_data(X_train,y_train,X_test,y_test)
```

```python
w = np.random.randn()
b = np.random.randn()
```

```python
w,b
```

```python
def loss_function(y_actual,y_pred):
    
    return np.sum((y_actual-y_pred)**2)/len(y_actual)
```

```python
y_out = X_test*w + b 
plot_data(X_train,y_train,X_test,y_test,predictions=y_out)
```

```python
lr = 0.0001 
epochs = 100 
train_loss_list , test_loss_list = [],[]
for epoch in range(epochs):
    train_preds = X_train*w + b 
    train_loss = loss_function(y_train,train_preds)
    
    gradientW = (-2/len(X_train))* np.sum((y_train-train_preds)*X_train)
    gradientb = (-2/len(X_train)*np.sum(y_train-train_preds))
    
    w = w - gradientW*lr
    b = b - gradientb*lr 
    
    test_preds = X_test*w + b 
    test_loss = loss_function(y_test,test_preds)
    
    train_loss_list.append(train_loss)
    test_loss_list.append(test_loss)
```

```python
test_loss_list
```

```python
train_loss_list
```

```python
y_pred = X_test*w + b 
```

```python
plot_data(X_train,y_train,X_test,y_test,predictions=y_pred)
```

```python
# loss vs epochs 

plt.plot(range(epochs),train_loss_list,label="Training Loss")
plt.plot(range(epochs),test_loss_list,label="Testing loss")
plt.xlabel("epochs")
plt.ylabel("train and test losses")
plt.legend()
plt.show()
```
