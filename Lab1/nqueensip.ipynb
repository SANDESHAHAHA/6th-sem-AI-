### N-Queens Problem

### Time Complexity: O(n!)

```python
def is_safe(board,x,y,n):
    # column test 
    for row in range(x):
        if board[row][y] == 'Q':
            return False
    
    # Upper diagonal test(left upper diagonally)
    r = x-1
    c = y-1
    while(r>=0 and c>=0):
        if(board[r][c]=='Q'):
            return False
        r-= 1 
        c-= 1
    
    # lower diagonal test(upper diagonally)
    r = x-1
    c = y+1
    while(r>=0 and c<n):
        if(board[r][c]=='Q'):
            return False 
        r-=1
        c+=1
    return True
```

### check from up to down

```python
n = 4
board = [['_']*4 for i in range(n)]
board[0][2] = 'Q'
print(is_safe(board,1,2,n))
```

```python
from pprint import pprint
n = int(input("Enter a number of Q's"))
board = [['_']*n for i in range(n)]
def nQueensSolver(board,row,n):
    if(row>=n):
        return True
    for col in range(n):
        if is_safe(board,row,col,n):
            board[row][col] = 'Q'
            if(nQueensSolver(board,row+1,n)):
                return True
            board[row][col] = '_'
    
    return False

if(nQueensSolver(board,0,n)):
    pprint(board)
    
else:
    pprint("No solution")
```
