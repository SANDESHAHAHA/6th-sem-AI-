```python


from collections import deque

grid=[
    ['S', '.', '.', 'X', '.'],
    ['.', 'X', '.', 'X', '.'],
    ['.', 'X', '.', '.', '.'],
    ['.', '.', 'X', 'X', '.'],
    ['.', '.', '.', '.', 'G']
]
rows=5
columns=5

def find_Positions(grid):
    start=goal=None

    for r in range(rows):
        for c in range(columns):
            if grid[r][c]=='S':
                start=(r,c)
            elif grid[r][c]=='G':
                goal=(r,c)

    return start,goal

def is_valid(r,c):
    return(
        0<=r<rows and
        0<=c<columns and
        grid[r][c]!='X'
    )

def get_neighbors(state):
    r,c=state

    moves=[
        (r-1,c,'UP'),
        (r+1,c,'DOWN'),
        (r,c-1,'LEFT'),
        (r,c+1,"RIGHT")
    ]

    neighbors=[]

    for R,C,action in moves:
        if is_valid(R,C):
            neighbors.append(((R,C),action))

    return neighbors

def robot(initial_state,goal_state):
    visited=set([initial_state])
    queue=deque([[(initial_state,'Start')]])
    while queue:
        path=queue.popleft()
        Current=path[-1][0]
        if Current==goal_state:
            print('Solution FOund!\n')

            for state,action in path:
                print(action)
                print(state)
                print()
            return
        for neighbour,action in get_neighbors(Current):
            if neighbour not in visited:
                visited.add(neighbour)
                new_path = path + [(neighbour, action)]
                queue.append(new_path)
    print("No Solution Found!")
start,goal=find_Positions(grid)
robot(start,goal)

```
